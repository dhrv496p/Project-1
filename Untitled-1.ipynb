```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading in the CSV File
survey_data = pd.read_csv('cleaned_majors_survey.csv')

# Displaying the first couple of piecees of information
survey_data.head(), survey_data.info()

# Creates a new column to differentiate betwween majors and non-majors
survey_data['is_major'] = survey_data['degree_program'].str.contains("computing", case=False, na=False)

# Calculating the interest levels by major status:
interest_level_before_applying = survey_data.groupby('is_major')['interest_level'].value_counts(normalize=True).unstack().fillna(0) * 100
```

```python
#Plots the results
plt.figure(figsize=(10, 6))
interest_level_before_applying.plot(kind='bar', stacked=True)
plt.title('Percentage fo Students Showing Interest Before applying by major status')
plt.xlabel('Is Major (Computing)')
plt.ylabel('Percentage')
plt.legend(title='Interest Level')
plt.show()
```

```python
#Filters for computing majors
course_counts = survey_data[survey_data['is_major']]['course'].value_counts().head(10)

#Plots the graph
plt.figure(figsize=(10, 6))
course_counts.plot(kind='barh')
plt.title('Top Influential Courses for Computing Majors')
plt.xlabel('Count')
plt.ylabel('Course')
plt.show()
```

```python
non_majors = survey_data[~survey_data['is_major']]
```

```python
#Calculating the percentage while keeping the switch in mind
considering_switch = non_majors['interest_level'].value_counts(normalize=True).loc[lambda x: x.index > 2].sum() * 100
print(f"Percentage of non-majors considering a switch to a computing major: {considering_switch:.2f}%")
```

```python
#Setting the satisfaction columns
satisfaction_columns = ['career_advancement', 'career_change']

#Calculating the satisfaction counts for each group
satisfaction_counts = survey_data.groupby('is_major')[satisfaction_columns].apply(lambda x: x.apply(lambda y: y.value_counts().get('Yes', 0), axis=0))

#Plotting the satisfaction Comparsion
plt.figure(figsize=(10, 6))
satisfaction_counts.plot(kind='bar', stacked=True)
plt.title('Satisfaction with IT Support (Using Career Advancement/Change as Proxies)')
plt.xlabel('Is Major (Computing)')
plt.ylabel('Count')
plt.legend(title='Satisfaction Factors')
plt.show()
```
